```python
from scipy.stats import levene
from scipy import stats 
```

```python
stat,p_value=levene(Brazil,England,France,Germany,Spain)
print("Stat:",stat,"P Value:",p_value)
```

```python
t_statistic,p_value=stats.f_oneway(Brazil,England,France,Germany,Spain)
print ("P Value:",p_value)
alpha=0.05
if p_value<alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
data['Total contribution score']=(data['Goals Scored']+ data['Assists']+data['Shots on Target']+data['Dribbling Success']+data['Aerial Duels Won']+data['Defensive Contribution']+data['Big Game Performance']+data['Consistency'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
data['Footedness']=encoder.fit_transform(data['Footedness'])
data['Marital Status']=encoder.fit_transform(data['Marital Status'])
```

```python
data.head()
```

```python
dummies = pd.get_dummies(data['Nationality'], dtype=int)
processed_df = pd.concat([data, dummies], axis=1)
processed_df = processed_df.drop('Nationality', axis=1)
processed_df.head()
```

```python
x=processed_df.drop(['Striker_ID'],axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,15):
    kmeans=KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(x)
    wcss_score=kmeans.inertia_
    wcss.append(wcss_score)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,15),wcss,marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusterss')
plt.ylabel('WCSS')
plt.show()
```

```python
final_km=KMeans(n_clusters=2)
final_km.fit(x)
labels=final_km.labels_
labels
```

```python
processed_df['Clusters']=labels
processed_df.head()
```

```python
round(processed_df.groupby('Clusters')[['Total contribution score']].mean(),2)
```

```python
mapping={0:'Best strikers',1:'Regular strikers'}
processed_df['Strikers types']=processed_df['Clusters'].map(mapping)
```

```python

processed_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_reg=processed_df.drop(['Striker_ID','Strikers types','Clusters'],axis=1)
y_reg=processed_df['Strikers types']
```

```python
scaler=StandardScaler()
scaled_x_reg=scaler.fit_transform(x_reg)
scaled_x_reg
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_reg_train,x_reg_test,y_reg_train,y_reg_test=train_test_split(scaled_x_reg,y_reg,test_size=0.2,random_state=42)
x_reg_train
```

```python
x_reg_train.shape
```

```python
x_reg_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
lgr_model=LogisticRegression()
lgr_model.fit(x_reg_train,y_reg_train)
y_lgr_pred=lgr_model.predict(x_reg_test)
accuracy_lgr=accuracy_score(y_reg_test,y_lgr_pred)
print(accuracy_lgr*100,'%')
```

```python
conf_matrix=confusion_matrix(y_reg_test,y_lgr_pred)
plt.figure(figsize=(12,6))
sns.heatmap(conf_matrix,annot=True,fmt="d",cmap="Blues")
plt.title("Confusion Matrix for the LGR model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
